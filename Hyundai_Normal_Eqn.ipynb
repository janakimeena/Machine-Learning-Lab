```python
class LR_Norm_Eqn:

  def __init__(self,m):
    # This is the feature / parameter (2x2) vector that will
    # contain my minimized values
    self.weights = np.zeros([m,1])    
  
  
  def predict(self, D, R): # retruns mean squared error
    temp = np.dot(D, self.weights) - R
    m = D.shape[0] # number of records
    return np.sum(np.power(temp, 2)) / (2*m)


  def fit(self,D,R):        
    
    D_transpose = np.transpose(D)
    
    D_trans_dot_D_Inv = np.linalg.inv(D_transpose.dot(D))

    self.weights = (D_trans_dot_D_Inv.dot(D_transpose)).dot(R)
    print("Dimension in fit",self.weights)
    return self.weights
```

```python
length = D.shape[1]
lin = LR_Norm_Eqn(length)
print('weights is ',lin.fit(D,r))
```
